```python
import pandas as pd 
import yfinance as yf
```

```python
# Download ticker from Yahoo Finance 
df_yahoo = yf.download ("AAPL", start = "2000-01-01", end = "2010-12-31")
df_yahoo
```

```python
# Download a list of multiple tickers 
df_yahoo2 = yf.download("MSFT AAPL", start="2000-01-01", end="2020-12-31")
df_yahoo2
```

```python
# Set to download only the adjusted price 
df_yahoo3 = yf.download("AAPL", start="2000-01-01", end="2020-12-31", auto_adjust = True)
df_yahoo3
```

```python
# Additionally download dividends and stock splits  
df_yahoo4 = yf.download("AAPL", start="2000-01-01", end="2020-12-31", actions = "inline")
df_yahoo4
```
